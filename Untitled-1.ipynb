```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lowprice,uppprice = remove_outlier(train['CoapplicantIncome'])
train['CoapplicantIncome']=np.where(train['CoapplicantIncome']>uppprice,uppprice,train['CoapplicantIncome'])
train['CoapplicantIncome']=np.where(train['CoapplicantIncome']<lowprice,lowprice,train['CoapplicantIncome'])
```

```python
data1 = pd.get_dummies(data)
data1.head()
```

```python
data_test1 = pd.get_dummies(data_test)
data_test1.head()
```

```python
id = test['Loan_ID']
id
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
```

```python
factors = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
```

```python
clf = DecisionTreeClassifier()
train_data = data1[factors]
test_data = data_test1[factors]
target = train.Loan_Status
```

```python
clf.fit(train_data, target)
y_pred = clf.predict(test_data)
```
